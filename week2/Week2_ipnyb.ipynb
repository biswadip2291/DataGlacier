**Hypothesis 3:"Cab companies experience different ride volumes on holidays and revenue patterns on holidays compared to non-holidays."**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from scipy.stats import f_oneway

# Create a dictionary mapping holiday dates to their names
holiday_names = {
    datetime(2016, 1, 1): 'New Year\'s Day',
    datetime(2016, 1, 18): 'Martin Luther King Jr. Day',
    datetime(2016, 2, 15): 'Washington\'s Birthday',
    datetime(2016, 5, 30): 'Memorial Day',
    datetime(2016, 7, 4): 'Independence Day',
    datetime(2016, 9, 5): 'Labor Day',
    datetime(2016, 10, 10): 'Columbus Day',
    datetime(2016, 11, 11): 'Veterans Day',
    datetime(2016, 11, 24): 'Thanksgiving Day',
    datetime(2016, 12, 25): 'Christmas Day',
    datetime(2017, 1, 1): 'New Year\'s Day',
    datetime(2017, 1, 16): 'Martin Luther King Jr. Day',
    datetime(2017, 2, 20): 'Washington\'s Birthday',
    datetime(2017, 5, 29): 'Memorial Day',
    datetime(2017, 7, 4): 'Independence Day',
    datetime(2017, 9, 4): 'Labor Day',
    datetime(2017, 10, 9): 'Columbus Day',
    datetime(2017, 11, 11): 'Veterans Day',
    datetime(2017, 11, 23): 'Thanksgiving Day',
    datetime(2017, 12, 25): 'Christmas Day',
    datetime(2018, 1, 1): 'New Year\'s Day',
    datetime(2018, 1, 15): 'Martin Luther King Jr. Day',
    datetime(2018, 2, 19): 'Washington\'s Birthday',
    datetime(2018, 5, 28): 'Memorial Day',
    datetime(2018, 7, 4): 'Independence Day',
    datetime(2018, 9, 3): 'Labor Day',
    datetime(2018, 10, 8): 'Columbus Day',
    datetime(2018, 11, 11): 'Veterans Day',
    datetime(2018, 11, 22): 'Thanksgiving Day',
    datetime(2018, 12, 25): 'Christmas Day',
}

holidays1 = list(holiday_names.keys())


master['Date of Travel'] = pd.to_datetime(master['Date of Travel'])
master['Is Holiday'] = master['Date of Travel'].isin(holidays1)

master['Holiday Name'] = master['Date of Travel'].apply(lambda x: holiday_names.get(x))

holiday_analysis = master.groupby(['Company', 'Is Holiday']).agg({
    'Transaction ID': 'count', 
    'Price Charged': 'mean'
}).reset_index()
holiday_analysis.columns = ['Company', 'Is Holiday', 'Number of Rides', 'Average Revenue per Ride']
holiday_data = holiday_analysis[holiday_analysis['Is Holiday'] == True]
non_holiday_data = holiday_analysis[holiday_analysis['Is Holiday'] == False]


rides_per_holiday = master[master['Is Holiday'] == True].groupby(['Holiday Name', 'Company']).agg({
    'Transaction ID': 'count'
}).reset_index()
rides_per_holiday.columns = ['Holiday Name', 'Company', 'Number of Rides']

rides_per_holiday = rides_per_holiday.sort_values(by=['Holiday Name', 'Company'])


pink_cab_holidays = master[(master['Company'] == 'Pink Cab') & (master['Is Holiday'] == True)]['Transaction ID']
yellow_cab_holidays = master[(master['Company'] == 'Yellow Cab') & (master['Is Holiday'] == True)]['Transaction ID']

# Perform ANOVA for number of rides on holidays for Pink Cab and Yellow Cab
anova_rides_results = f_oneway(pink_cab_holidays, yellow_cab_holidays)

# Print ANOVA results
print("ANOVA results for Number of Rides during Holidays:")
print("F-statistic:", anova_rides_results.statistic)
print("p-value:", anova_rides_results.pvalue)

pink_cab_holidays_rev = master[(master['Company'] == 'Pink Cab') & (master['Is Holiday'] == True)]['Price Charged']
pink_cab_non_holidays_rev = master[(master['Company'] == 'Pink Cab') & (master['Is Holiday'] == False)]['Price Charged']

yellow_cab_holidays_rev = master[(master['Company'] == 'Yellow Cab') & (master['Is Holiday'] == True)]['Price Charged']
yellow_cab_non_holidays_rev = master[(master['Company'] == 'Yellow Cab') & (master['Is Holiday'] == False)]['Price Charged']

# Perform ANOVA for average revenue per ride
anova_revenue_results_pink = f_oneway(pink_cab_holidays_rev, pink_cab_non_holidays_rev)
anova_revenue_results_yellow = f_oneway(yellow_cab_holidays_rev, yellow_cab_non_holidays_rev)

print("\nANOVA results for Average Revenue per Ride:")
print("Pink Cab - Holidays vs Non-Holidays:")
print("F-statistic:", anova_revenue_results_pink.statistic)
print("p-value:", anova_revenue_results_pink.pvalue)

print("\nYellow Cab - Holidays vs Non-Holidays:")
print("F-statistic:", anova_revenue_results_yellow.statistic)
print("p-value:", anova_revenue_results_yellow.pvalue)


# Plotting the number of rides on holidays by company


fig, ax = plt.subplots(figsize=(18, 10))

companies = rides_per_holiday['Company'].unique()
holidays = rides_per_holiday['Holiday Name'].unique()
x = np.arange(len(holidays))

bar_width = 0.35
for i, company in enumerate(companies):
    company_data = rides_per_holiday[rides_per_holiday['Company'] == company]
    company_rides = []
    for holiday in holidays:
        rides = company_data[company_data['Holiday Name'] == holiday]['Number of Rides']
        if not rides.empty:
            company_rides.append(rides.values[0])
        else:
            company_rides.append(0)
    bar_positions = x + (i * bar_width)
    ax.bar(bar_positions, company_rides, width=bar_width, label=company)

ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(holidays, rotation=45, ha='right')
ax.set_title('Number of Rides on Holidays by Company')
ax.set_ylabel('Number of Rides')
ax.legend()

plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(18, 10))

bar_width = 0.35
x = np.arange(len(holiday_data))

ax.bar(x - bar_width/2, holiday_data['Average Revenue per Ride'], bar_width, label='Holidays', alpha=0.7, color='orange')
ax.bar(x + bar_width/2, non_holiday_data['Average Revenue per Ride'], bar_width, label='Non-Holidays', alpha=0.7, color='blue')
ax.set_xticks(x)
ax.set_xticklabels(holiday_data['Company'], rotation=90)
ax.set_title('Average Revenue per Ride during Holidays vs Non-Holidays')
ax.set_ylabel('Average Revenue per Ride')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
from statsmodels.stats.multicomp import pairwise_tukeyhsd


analysis = master.groupby(['Company', 'Age Group']).agg({
    'Transaction ID': 'count', 
    'profit': 'sum'
}).reset_index()

# Perform ANOVA for Number of Rides
anova_rides = f_oneway(
    analysis[analysis['Company'] == 'Pink Cab']['Transaction ID'],
    analysis[analysis['Company'] == 'Yellow Cab']['Transaction ID']
)

print("ANOVA Results for Number of Rides:")
print(anova_rides)

# Perform ANOVA for Total Profit
anova_profit = f_oneway(
    analysis[analysis['Company'] == 'Pink Cab']['profit'],
    analysis[analysis['Company'] == 'Yellow Cab']['profit']
)

print("\nANOVA Results for Total Profit:")
print(anova_profit)

if anova_rides.pvalue < 0.05:
    tukey_rides = pairwise_tukeyhsd(
        analysis['Transaction ID'], analysis['Company']
    )
    print("\nTukey HSD Test Results for Number of Rides:")
    print(tukey_rides)

if anova_profit.pvalue < 0.05:
    tukey_profit = pairwise_tukeyhsd(
        analysis['profit'], analysis['Company']
    )
    print("\nTukey HSD Test Results for Total Profit:")
    print(tukey_profit)
```

**Hypothesis:6"Does margin proportionally increase with an increase in the number of customers?**

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt



company_summary = master.groupby('Company').agg({
    'Customer ID': 'nunique',       
    'Price Charged': 'sum',         
    'profit': 'sum'                 
}).reset_index()

company_summary['Margin'] = (company_summary['profit'] / company_summary['Price Charged']) * 100

company_summary.columns = ['Company', 'Total Customers', 'Total Revenue', 'Total Profit', 'Margin']

# Perform ANOVA to test if there is a significant difference in margins between companies
pink_margin = master[master['Company'] == 'Pink Cab']['profit']
yellow_margin = master[master['Company'] == 'Yellow Cab']['profit']

f_stat, p_value = stats.f_oneway(pink_margin, yellow_margin)

print(f'ANOVA F-statistic: {f_stat}')
print(f'ANOVA p-value: {p_value}')

fig, axes = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

bars1 = axes[0].bar(company_summary['Company'], company_summary['Total Customers'], color=['pink', 'yellow'])
axes[0].set_title('Total Customers by Company')
axes[0].set_ylabel('Total Customers')

for bar in bars1:
    height = bar.get_height()
    axes[0].annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

bars2 = axes[1].bar(company_summary['Company'], company_summary['Margin'], color=['pink', 'yellow'])
axes[1].set_title('Margin by Company')
axes[1].set_ylabel('Margin (%)')
axes[1].set_xlabel('Company')

for bar in bars2:
    height = bar.get_height()
    axes[1].annotate(f'{height:.2f}%', xy=(bar.get_x() + bar.get_width() / 2, height),
                     xytext=(0, 3), textcoords='offset points', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
